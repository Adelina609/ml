```python
import plotly.graph_objects as plt
import numpy as np

n = 100
x = [np.random.randint(0, n) for i  in range(n)]
y = [np.random.randint(0, n) for i  in range(n)]
z = [np.random.randint(0, n) for i  in range(n)]
points = []
for i in range(n):
    points.append([x[i], y[i], z[i]])
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0).fit(points)
clusters = kmeans.labels_
colors = ['yellow'] * n

for i in range(n):
    if clusters[i] == 1:
        colors[i] = 'deeppink'

figure = plt.Figure(data=[plt.Scatter3d(x=x, y=y, z=z, mode='markers', marker=dict(color=colors))])
x_new = np.random.randint(0, n)
y_new = np.random.randint(0, n)
z_new = np.random.randint(0, n)
points.append([x_new, y_new, z_new])
```

```python
from sklearn.svm import SVC
figure.add_trace(plt.Scatter3d(x=x, y=y, z=z, mode='markers', marker=dict(color=colors)))

svc = SVC(kernel='linear')
svc.fit(points[:n], clusters)

tmp = np.linspace(0, 100, 50)
xx, yx = np.meshgrid(tmp, tmp)
zz = (-svc.intercept_[0] - svc.coef_[0][0] * xx - svc.coef_[0][1] * yx) / svc.coef_[0][2]

figure.add_trace(plt.Surface(x=xx, y=yx, z=zz))
figure.show()
```
